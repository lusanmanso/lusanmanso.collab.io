## Question 1/6

The dot product between two vectors $v1$ e $v2$ of size $n$ is defined as:

$$\sum_{i=1}^n v1_i . v2_i $$

Create a `for` loop necessary to compute the dot product between any two vectors of the same size.



````python
v1 = [3,4,5,7,9]
v2 = [1,5,6,7,8]

def prodAlterno(a,b):
  resultado = 0
  """ Multiplicar alternadamente los elementos en la misma posición de cada funcion
  """
  # zip[a, b] Solo asigna a dos vectores distintos
  for el_v1, el_v2 in list(zip(v1, v2)):
    resultado += el_v1 * el_v2

  return resultado

print(prodAlterno(v1, v2))
````

## Question 4/6

The norm of a vector (its magnitude) is defined as:

$$ \lVert x \rVert = \sqrt{\sum_{i=1}^n x_i^2}$$

Write a for loop that computes a norm of a given vector $v$.

````python
v1 = [1, 2, 3, 4, 5]

def norm_vect(v):
  sum_v = 0
  for el in v:
    sum_v += el**2

  return sum_v**0.5

print(norm_vect(v1))

# return sum(el**2 in el for v)]**0.5
````
